This notebook explores the Capstan equation 

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import math
```

```python
radius = 1.0  # Radius of round capstan
mu = 0.5 # coefficient of static friction
pi = math.pi
rtpi = math.sqrt(math.pi)
```

```python
# Draw a circle representing a rod or dowel that the rope is wrapped around
circN = 1001
thetas = np.linspace(0,2*pi,circN)
circX = np.zeros_like(thetas)
circY = np.zeros_like(thetas)
for i in range(circN):
    circX[i] = math.cos(thetas[i])
    circY[i] = math.sin(thetas[i])
```

```python
fig0=plt.figure()
ax0 = fig0.add_subplot(1,1,1)
plt.plot(circX,circY,"k-",lw=12)
ax0.set_aspect('equal', adjustable='box')
plt.show()
```
